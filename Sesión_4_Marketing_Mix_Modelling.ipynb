# ¿Qué es Marketing Mix Modelling?

El Marketing Mix Modeling abarca el método de análisis estadístico, como la regresión en las ventas y los datos de marketing para estimar el impacto de varias tácticas de marketing en las ventas y luego predecir el impacto de futuros conjuntos de tácticas. También se utiliza para optimizar la combinación de publicidad y las tácticas promocionales con respecto a los ingresos por ventas o ganancias. Hoy, la mayoría de las marcas y empresas han adoptado y están adoptando estas metodologías de medición avanzadas para determinar el verdadero impacto de todas las interacciones en todos los dispositivos, canales y tácticas de campaña.

Hoy en día, la mayoría de las empresas tienen una base sólida de modelos econométricos con estrategias ligeramente diferentes según sus requisitos, productos, competencia externa, etc.

## Entendiendo cada palabra de "Marketing Mix Modelling"

> Al romper la frase, obtenemos tres palabras, es decir, marketing, mezcla y modelado. Exploremos estos términos.
1. Marketing: este término corresponde a la comprensión del mercado.
2. Mezclar: Mezclar aquí se refiere a la importancia que se le debe dar a cada P en el 4 P (Producto, precio, lugar y promoción)
3. Modelado: el modelado aquí se refiere a la optimización y construcción del modelo.

# Diseño de la combinación de marketing correcta:

El paso más creativo y desafiante en el marketing es diseñar la combinación de marketing adecuada. La combinación de marketing es una colección específica de acciones e instrumentos asociados empleados por una organización para estimular la aceptación de sus ideas, productos y servicios.

## Oferta final al cliente:

La oferta final hecha al cliente o consumidor depende del gerente de marketing de la empresa.
donde se siguen los siguientes cuatro pasos.
1. La empresa (o el gerente de marketing elige el producto para satisfacer las demandas del mercado).
2. El canal de distribución correcto se utiliza para hacer que el producto esté disponible.
3. Se ofrece al cliente una promoción llamativa para los productos.
4. Además de la promoción, se fija el precio apropiado acordado por la empresa y el cliente.

### Nota:
1. Sin embargo, el término 'marketing mix' no implica que los elementos 4P representen opciones.
2. No son compensaciones, pero son cuestiones fundamentales de marketing que siempre deben abordarse.
3. Son las acciones fundamentales que requiere el marketing, ya sea determinado explícitamente o por defecto.



## Optimizando la mezcla de marketing:
> Al ofrecer el producto con una combinación de 4 P, la popularidad del producto y los ingresos por ventas de la empresa que vende ese producto aumentarán gradualmente.
Hay 2 optimizaciones que se pueden implementar.
1. Optimización estratégica:
   > Esto aparece cuando se realizan grandes cambios en el peso dado a una P particular o un grupo de P's. Por ejemplo, un gran cambio en el precio de S/.500 a S/.1000 podría tener un profundo efecto en los ingresos por ventas. Este tipo de cambios cae en esta categoría.
2. Optimización táctica:
   > Esto aparece cuando se hacen pequeños cambios en el peso dado a una P particular o un grupo de P's. Por ejemplo, considerando el ejemplo anterior para la Optimización estratégica, si el Precio cambia de S/.500 a S/.499, entonces este tipo de cambios se encuadra en la Optimización táctica.

# Fortalezas y debilidades de MMM:

## Fortalezas:
> Un modelo de marketing mix implementado con éxito puede proporcionar los siguientes beneficios:
  1. Reflexiona sobre las decisiones de marketing tomadas en el pasado, calculando el retorno de la inversión (ROI).
  2. Ayuda a optimizar el mix de marketing para las empresas al pronosticar el impacto probable de los cambios en varias variables del mix de marketing.
  3. Ayuda a encontrar los cambios en el desempeño del negocio al aislar el impacto de factores internos y externos.

## Debilidad:
>
1. El enfoque en las ventas a corto plazo puede subestimar significativamente la importancia de las actividades de creación de capital a largo plazo
2. MMM, utilizado para la optimización de la mezcla de medios, tiene un sesgo a favor de los medios dependientes del tiempo (como la televisión) frente a los medios menos dependientes del tiempo (como los anuncios que aparecen en revistas mensuales); Los sesgos también pueden ocurrir cuando se comparan medios de amplia base versus medios demográficamente dirigidos.

 **Trabajando un MMM**

Comprendamos el funcionamiento de MMM con un ejemplo.
Básicamente, los MMM se analizan e implementan con algoritmos de aprendizaje estadísticos como la regresión.

Se utilizan muchos tipos de regresión en MMM para una mejor precisión y rendimiento. Algunos de ellos son regresión lineal, regresión multivariada, etc.

Primero, vamos a importar las bibliotecas y leer los datos, como de costumbre.

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
#Importamos la data
url = 'https://raw.githubusercontent.com/JBrianAlicorp/Business-Analytics/master/Data_Marketing_Mix_Modelling.csv'
df = pd.read_csv(url,encoding='latin1')
```

A continuación, veremos las variables en el conjunto de datos y entenderemos con qué estamos trabajando.

```
df.head()
```

Esta data consiste en gastos de marketing en televisión, radio y periódicos, así como las ventas correspondientes por período.

```
print(df.columns)
df.describe()
```

En mi base cuento con datos de gastos de marketing de TV, radio y periódico y el total de ventas mensual.

Al instante, puede ver que la variable Innominado: 0 es esencialmente un índice que comienza en 1, por lo que lo vamos a eliminar.

```
df = df.copy().drop(['Unnamed: 0'],axis=1)
df = df.copy().drop(['Periodo'],axis=1)
df.head()
```

**Análisis de datos exploratorio**

Lo primero que siempre me gusta hacer es crear una matriz de correlación porque me permite comprender mejor las relaciones entre mis variables de un vistazo.

```
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

Inmediatamente, podemos ver que hay una fuerte correlación entre TV y ventas (0.78), una correlación moderada entre radio y ventas (0.58), y una correlación débil entre periódico y ventas (0.23). Todavía es demasiado pronto para concluir algo, pero es bueno tenerlo en cuenta para seguir adelante.
De manera similar a la matriz de correlación, quiero crear un diagrama de pares de mis variables para poder comprender aún más las relaciones entre mis variables.

```
sns.pairplot(df)
```

Esto parece estar en línea con la matriz de correlación, ya que parece haber una fuerte relación entre la televisión y las ventas, menos para la radio y aún menos para los periódicos.

**Importancia de las variables**

La importancia de la característica le permite determinar cuán "importante" es cada variable de entrada para predecir la variable de salida. Una característica es importante si mezclar sus valores aumenta el error del modelo porque esto significa que el modelo se basó en la característica para la predicción.
Vamos a crear rápidamente un modelo de bosque aleatorio para que podamos determinar la importancia de cada característica.

```
# Setting X and y variables
X = df.loc[:, df.columns != 'sales']
y = df['sales']
# Building Random Forest model
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=0)
model = RandomForestRegressor(random_state=1)
model.fit(X_train, y_train)
pred = model.predict(X_test)
# Visualizing Feature Importance
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(25).plot(kind='barh',figsize=(10,10))
```

Parece haber un patrón, donde la televisión es lo más importante, seguida de la radio, dejando el periódico al final. Realmente construyamos nuestro modelo de regresión OLS.

```
import statsmodels.formula.api as sm
model = sm.ols(formula="sales~TV+radio+newspaper", data=df).fit()
print(model.summary())
```

.summary () nos proporciona una gran cantidad de información sobre nuestro modelo. Voy a señalar dos cosas principales que son más útiles para nosotros en esto:

El R-cuadrado es 0.896, lo que significa que casi 90 de todas las variaciones en nuestros datos pueden explicarse por nuestro modelo, ¡lo cual es bastante bueno!

Los valores p para TV y radio son inferiores a 0,000, pero el valor p para periódicos es 0,86, lo que indica que el gasto en periódicos no tiene un impacto significativo en las ventas.
A continuación, grafiquemos los valores de ventas pronosticados con los valores de ventas reales para ver visualmente cómo funciona nuestro modelo:

```
# Defining Actual and Predicted values
y_pred = model.predict()
labels = df['sales']
df_temp = pd.DataFrame({'Actual': labels, 'Predicted':y_pred})
df_temp.head()
# Creating Line Graph
from matplotlib.pyplot import figure
figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
y1 = df_temp['Actual']
y2 = df_temp['Predicted']
plt.plot(y1, label = 'Actual')
plt.plot(y2, label = 'Predicted')
plt.legend()
plt.show()
```

¡No está mal! Parece que este modelo hace un buen trabajo al predecir las ventas dados los gastos de marketing de televisión, de radio y de periódicos.

En la práctica, es probable que desee considerar más variables que afecten las ventas, incluidas, entre otras, las siguientes:

**Estacionalidad** : casi siempre ocurre que las ventas de la empresa son estacionales. Por ejemplo, las ventas de una compañía de snowboard serían mucho más altas durante el invierno que en el verano. En la práctica, querrás incluir una variable para tener en cuenta la estacionalidad.

**Efectos de arrastre** : el impacto del marketing no suele ser inmediato. En muchos casos, los consumidores necesitan tiempo para pensar en sus decisiones de compra después de ver anuncios. Los efectos de transferencia representan el lapso de tiempo entre el momento en que los consumidores están expuestos a un anuncio y su respuesta al anuncio.

**Ventas base vs ventas incrementales**: no todas las ventas se atribuyen al marketing. Si una empresa no gastara absolutamente nada en marketing y todavía hiciera ventas, esto se llamaría su base de ventas . Por lo tanto, para ir un paso más allá, podría intentar modelar el gasto publicitario en ventas incrementales en lugar de las ventas totales.
